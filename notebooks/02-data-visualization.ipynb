# Statistical methods to interpretate results for one-dimensional datasets

**One-dimensional dataset** represent one value/result of the specific type, and can be visualized on one-axix graph.

<br>

One dimensional data can be shown as an array of values where `X.size = n`. $$X = \big< x_0, x_1, x_2, ..., x_{n-1} \big>, \quad X = \big< X[0], X[1], X[2], ..., X[n-1] \big>$$

For example:
* When examining the distances covered by runners during a 5-minute run, our data will consist of the distances they ran. $$Distances = \big< 1078, 896, 1196, 1009, 1078, 1096, 923 \big>$$
* When studying the relationship between height and body weight in humans, one-dimensional data would separately represent height and weight. $$Weights = \big< 60, 76, 99, 48 \big>$$ $$Heights = \big< 177, 189, 197, 165 \big>$$ If we were to create a database combining the height and weight of each individual, we would then be dealing with **two-dimensional data**. $$Weights\_Heights = \big< (60, 177), (76, 189), (99, 197), (48, 165) \big>$$

### Methods for data occurence

We can calculate some statistical values to visualize the data occurence.

* **Mean**: Represents the central value of the dataset. $$Mean(X) = \frac{\displaystyle\sum_{i=0}^{n-1}{X[i]}}{\displaystyle n}$$ 

* **Variance**: Indicates how much the data values differ from the mean. The bigger variance, the more spread out the data are. $$Variance(X) = \frac{\displaystyle\sum_{i=0}^{n-1}{(X[i] - Mean(X))^2}}{\displaystyle n}$$

<br>

* **Standard Deviation**: Provides an easy-to-interpret measure of dispersion expressed in the same unit as the data. $$Standard\_Deviation(X) = Std(X) = \sqrt{Variance(X)}$$

```python
def mean(data: list) -> float:
    """Calculates the mean of the dataset containing numbers (integers or/and floats).
       
       Mean represents the central value of the dataset."""
    
    counter = 0

    for datum in data:
        counter = counter + datum

    return counter /len(data)


def variance(data: list) -> float:
    """Calculates the variance of the dataset containing numbers (integers or/and floats).
       
       Variance indicates how much the data values differ from the mean. The bigger variance, the more spread out the data are."""
    
    counter = 0
    data_mean = mean(data)

    for datum in data:
        counter = counter + (datum - data_mean) ** 2

    return counter /len(data)


def standard_deviation(data: list) -> float:
    """Calculates the standard deviation of the dataset containing numbers (integers or/and floats).
    
       Standard Deviation provides an easy-to-interpret measure of dispersion expressed in the same unit as the data."""
    
    return variance(data) ** 0.5


def std(data: list) -> float:
    """Calculates the standard deviation of the dataset containing numbers (integers or/and floats).
    
       Standard Deviation provides an easy-to-interpret measure of dispersion expressed in the same unit as the data."""
    
    return standard_deviation(data)
```

To visualize the dataset, I will use the `matplotlib` library to generate a plot. 

**Blue points** represent the values in the dataset. To make them easy to visualize, I positioned them along the $OX$ axis. For this purpose, I created an additional dataset that assigns a value of zero to each data point, ensuring the data is nicely displayed along the $OX$ axis.

A **dashed red line** indicates the central value of the dataset, which is the mean. The standard deviation should be interpreted in the context of the mean, as it indicates the spread of the data around it. Therefore, on the plot, you will see **two dashed green lines** representing the deviation from the mean. It's hard to miss that the highest density of data occurs between these green lines. Their distances were calculated using formula \
 $Mean(X) + Std(X)$ and $Mean(X) - Std(X)$.

```python
ordinal_numbers = [ 0 for datum in data ]

plt.scatter(
    x = data,
    y = ordinal_numbers,
)

data_mean = mean(data)
data_variance = variance(data)
data_std = std(data)

plt.axvline(
    x = data_mean,
    color = 'red',
    linestyle = '--',
    label = f'Mean = {np.round(data_mean, 3)}'
)

plt.axvline(
    x = data_mean + data_std,
    color = 'green',
    linestyle = '--',
    label = f'Std = {np.round(data_std, 3)}'
)

plt.axvline(
    x = data_mean - data_std,
    color = 'green',
    linestyle = '--',
)

plt.ylim([-0.05, 0.05])
plt.title('Generated numbers')
plt.legend()
plt.show()
```

### Functions Useful for Visualization.

* `plt.scatter(x, y, ...)`: Draws a scatter plot with points specified by the coordinates `x` and `y`.

* `plt.plot(x, y, ...)`: Draws a line plot, connecting points specified by `x` and `y` with a line.

* `plt.axhline(y, ...)`: Adds a horizontal line to the plot at height `y`.

* `plt.axvline(x, ...)`: Adds a vertical line to the plot at position `x`.

### Error bar

An error bar is a graphical representation of the variability of data, often used in scientific and statistical plots to indicate the uncertainty or error in a measurement. Error bars provide a way to visualize the range within which the true value of a data point is expected to lie.

The central point represents the mean value, while the ends of the arms represent the standard deviation.

<br>

**Horizontal error bar**: `plt.errorbar(x = means, y = labels, xerr = standard_deviations)`

**Vertical error bar**: `plt.errorbar(x = labels, y = means, yerr = standard_deviations)`

<br>

The most popular form of error bar representation.

`plt.errorbar(x, y, xerr, yerr, fmt = 'o', capsize = 5, color = 'black')`

```python
ordinal_numbers = [ 0 for datum in data ]

plt.scatter(
    x = data,
    y = ordinal_numbers,
)

data_mean = mean(data)
data_variance = variance(data)
data_std = std(data)

plt.errorbar(
    x = [data_mean],
    y = [0.01],
    xerr = [data_std],
    fmt = 'o',
    capsize = 8,
    color = 'black',
    label = f'Mean = {np.round(data_mean, 3)}, Std = {np.round(data_std, 3)}'
)

plt.ylim([-0.05, 0.05])
plt.title('Generated numbers')
plt.legend()
plt.show()
```

### Histogram

Histogram is a bar chart that represents the frequency of occurrences of values in a dataset within specified intervals.

For example, given a dataset containing numbers from 0 to 9 and 6 intervals, the bars on the $OY$ axis represent the number of values falling within the intervals $[0, 1.5], [1.5, 3], [3, 4.5], [4.5, 6], [6, 7.5], [7.5, 9]$.

For this purpose, I created a completely fictional dataset representing the results of a long jump tournament.

```python
# Randomly generated dataset by my hand
distances = [8.79, 3.91, 6.2, 3.22, 7.24, 4.11, 1.7, 7.89, 5.29]
```

To create a histogram, the `plt.hist(x, ...)` function should be used. To better illustrate the operation of the histogram, I marked all the dataset values on the line $y=\frac{1}{2}$ with red dots.

```python
ordinal_numbers = [ 0.5 for i in distances ]

plt.hist(distances)

plt.scatter(
    distances,
    ordinal_numbers,
    color = 'red'
)

plt.show()
```

The `plt.hist()` function automatically selects the best number of intervals, known as BINS. However, sometimes we may need to manually set their number. To do this, the `bins: int` parameter should be manually specified.

Let $bins = 2$.

```python
plt.hist(
    x = distances,
    bins = 2
)

plt.scatter(
    distances,
    ordinal_numbers,
    color = 'red'
)

plt.show()
```

Let $bins = 3$.

```python
plt.hist(
    x = distances,
    bins = 3
)

plt.scatter(
    distances,
    ordinal_numbers,
    color = 'red'
)

plt.show()
```

Let $bins = 5$.

```python
plt.hist(
    x = distances,
    bins = 5
)

plt.scatter(
    distances,
    ordinal_numbers,
    color = 'red'
)

plt.show()
```

However, caution is required, as too few or too many intervals may provide little insight into the dataset being analyzed.

In the following case, the consequence of having too many intervals is the uselessness of the histogram. It only shows the locations where the values occur in the dataset.

Let $bins = 100$.

```python
plt.hist(
    x = distances,
    bins = 100
)

plt.scatter(
    distances,
    ordinal_numbers,
    color = 'red'
)

plt.show()
```

The set of distances is small enough that using a histogram might not make much sense, although it doesn't mean it's not useful. For example, here is the histogram for the previous dataset `data`, where the distribution of the data is more easily interpreted.

Combined with the other visualization methods, the entire set of techniques becomes a powerful tool for analysis.

```python
ordinal_numbers = [ 5 for i in data ]

plt.hist(
    x = data
)

plt.scatter(
    data,
    ordinal_numbers,
    color = 'red'
)

plt.errorbar(
    x = [data_mean],
    y = [30],
    xerr = [data_std],
    fmt = 'o',
    capsize = 8,
    color = 'black',
    label = f'Mean = {np.round(data_mean, 3)}, Std = {np.round(data_std, 3)}'
)

plt.show()
```
